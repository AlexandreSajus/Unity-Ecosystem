```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
suffix = '_exp1.txt'
FOX_POPULATION_PATH = 'data/foxes' + suffix
FOX_SPEED_PATH = 'data/fox_speed' + suffix
RABBIT_POPULATION_PATH = 'data/rabbits' + suffix
RABBIT_SPEED_PATH = 'data/rabbit_speed' + suffix
paths = [FOX_POPULATION_PATH, FOX_SPEED_PATH, RABBIT_POPULATION_PATH, RABBIT_SPEED_PATH]
```

```python
# Replace ',' with '.' in the files
for path in paths:
    with open(path, 'r') as f:
        filedata = f.read()
    filedata = filedata.replace(',', '.')
    with open(path, 'w') as f:
        f.write(filedata)
```

```python
# Load the data
fox_population = np.loadtxt(FOX_POPULATION_PATH)
fox_speed = np.loadtxt(FOX_SPEED_PATH)
rabbit_population = np.loadtxt(RABBIT_POPULATION_PATH)
rabbit_speed = np.loadtxt(RABBIT_SPEED_PATH)
```

```python
# Plot the population data
plt.plot(fox_population, label='Foxes')
plt.plot(rabbit_population, label='Rabbits')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.show()
```

```python
# Plot the speed data
plt.plot(fox_speed, label='Foxes')
plt.plot(rabbit_speed, label='Rabbits')
plt.xlabel('Time')
plt.ylabel('Average Speed')
plt.legend()
plt.show()
```

```python
# Zoom factor
xlim = 30
```

```python
# Plot the population data zoomed
plt.plot(fox_population[:xlim], label='Foxes')
plt.plot(rabbit_population[:xlim], label='Rabbits')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.show()
```

```python
# Plot the speed data zoomed
plt.plot(fox_speed[:xlim], label='Foxes')
plt.plot(rabbit_speed[:xlim], label='Rabbits')
plt.xlabel('Time')
plt.ylabel('Average Speed')
plt.legend()
plt.show()
```

```python
# Plot two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.plot(fox_population[:xlim], label='Lions')
ax1.plot(rabbit_population[:xlim], label='Chickens')
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Population')
ax1.legend()
ax2.plot(fox_speed[:xlim], label='Lions')
ax2.plot(rabbit_speed[:xlim], label='Chickens')
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Average Speed (m/s)')
ax2.legend()
# Add title
fig.suptitle('Population and Speed of Lions and Chickens over Time')
plt.show()
```
